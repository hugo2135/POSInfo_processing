```python
from plotly import graph_objects as go
from transferPOS_2_anything import Single_Axis as toolbox
import numpy as np

tool = toolbox()
data = tool.read_measure_file("F10000_all_default_no_filter.csv")
```

```python
posf = data['CH1:POS3D']
vibr = tool.window(posf, [300.0005, 300.0038])
tool.time_interval = 0.001
vibr_dB, vibr_freq = tool.fanuc_rfft(vibr)
vibr_dB = vibr_dB[1:]
vibr_freq = vibr_freq[1:]
```

```python
go.Figure(data=go.Line(y=vibr, x=[x for x in range(len(vibr))]))
```

```python
from scipy.signal import find_peaks
peaks, _ = find_peaks(vibr_dB)
```

```python
fig = go.Figure()
fig.add_trace(go.Line(y=vibr_dB, x=vibr_freq))
fig.add_trace(go.Scatter(y=vibr_dB[peaks], x=vibr_freq[peaks], mode="markers"))
```
